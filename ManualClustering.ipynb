```python
import pandas as pd

data = pd.read_csv('data//1_preprocessed//preprocessed_dataset.csv')

data5clusters = data.copy()
data10clusters = data.copy()
data15clusters = data.copy()
data20clusters = data.copy()
data25clusters = data.copy()
data30clusters = data.copy()

# apply cluster mapping to the 'track_genre' column to create a new 'cluster' column
def map_genre_to_cluster5(genre):
    for cluster_name, genres in clusters_5.items():
        if genre in genres:
            return cluster_name
    return "Unknown Cluster"

def map_genre_to_cluster10(genre):
    for cluster_name, genres in clusters_10.items():
        if genre in genres:
            return cluster_name
    return "Unknown Cluster"

def map_genre_to_cluster15(genre):
    for cluster_name, genres in clusters_15.items():
        if genre in genres:
            return cluster_name
    return "Unknown Cluster"

def map_genre_to_cluster20(genre):
    for cluster_name, genres in clusters_20.items():
        if genre in genres:
            return cluster_name
    return "Unknown Cluster"

def map_genre_to_cluster25(genre):
    for cluster_name, genres in clusters_25.items():
        if genre in genres:
            return cluster_name
    return "Unknown Cluster"

def map_genre_to_cluster30(genre):
    for cluster_name, genres in clusters_30.items():
        if genre in genres:
            return cluster_name
    return "Unknown Cluster"

data5clusters['cluster'] = data5clusters['track_genre'].apply(map_genre_to_cluster5)
data10clusters['cluster'] = data10clusters['track_genre'].apply(map_genre_to_cluster10)
data15clusters['cluster'] = data15clusters['track_genre'].apply(map_genre_to_cluster15)
data20clusters['cluster'] = data20clusters['track_genre'].apply(map_genre_to_cluster20)
data25clusters['cluster'] = data25clusters['track_genre'].apply(map_genre_to_cluster25)
data30clusters['cluster'] = data30clusters['track_genre'].apply(map_genre_to_cluster30)

# Check for any "Unknown Cluster" entries to ensure all genres were mapped
unknown5 = data5clusters[data5clusters['cluster'] == "Unknown Cluster"]['track_genre'].unique()
unknown10 = data10clusters[data10clusters['cluster'] == "Unknown Cluster"]['track_genre'].unique()
unknown_15 = data15clusters[data15clusters['cluster'] == "Unknown Cluster"]['track_genre'].unique()
unknown_20 = data20clusters[data20clusters['cluster'] == "Unknown Cluster"]['track_genre'].unique()
unknown_25 = data25clusters[data25clusters['cluster'] == "Unknown Cluster"]['track_genre'].unique()
unknown_30 = data30clusters[data30clusters['cluster'] == "Unknown Cluster"]['track_genre'].unique()
print("Unknown genres in 5 clusters:", unknown5)
print("Unknown genres in 10 clusters:", unknown10)
print("Unknown genres in 15 clusters:", unknown_15)
print("Unknown genres in 20 clusters:", unknown_20)
print("Unknown genres in 25 clusters:", unknown_25)
print("Unknown genres in 30 clusters:", unknown_30)
```
