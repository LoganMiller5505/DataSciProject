```python
import pandas as pd

data = pd.read_csv('data//1_preprocessed//preprocessed_dataset.csv')

# apply cluster mapping to the 'track_genre' column to create a new 'cluster' column
def map_genre_to_cluster5(genre):
    for cluster_name, genres in clusters_5.items():
        if genre in genres:
            return cluster_name
    return "Unknown Cluster"

data['cluster'] = data['track_genre'].apply(map_genre_to_cluster5)

unknown5 = data[data['cluster'] == "Unknown Cluster"]['track_genre'].unique()
```

# Supergenre Relationships

## Audio Feature Averages for Supergenres

```python
import matplotlib.pyplot as plt

features_to_plot = ['danceability', 'energy', 'acousticness', 'speechiness']
short_names = {
    'Rock, Metal & Punk Universe':               'Rock/Metal',
    'Electronic & Dance Music':                  'Electronic',
    'Pop, Entertainment & Global Pop':           'Pop',
    'Latin, Reggae & World Cultural Traditions': 'Latin/Reggae',
    'Roots, Classical & Introspective Music':    'Roots/Classical',
}

cluster_means = data.groupby('cluster')[features_to_plot].mean()
cluster_means.index = cluster_means.index.map(short_names)

# Assign consistent colors by cluster identity across all panels
cluster_color_map = {short_names[c]: plt.cm.tab10.colors[i] for i, c in enumerate(clusters_5.keys())}

fig, axes = plt.subplots(1, 4, figsize=(18, 5))
for ax, feature in zip(axes, features_to_plot):
    sorted_means = cluster_means[feature].sort_values()
    colors = [cluster_color_map[name] for name in sorted_means.index]
    ax.barh(sorted_means.index, sorted_means.values, color=colors)
    ax.axvline(data[feature].mean(), color='black', linestyle='--', linewidth=1, label='Dataset mean')
    ax.set_title(f'{feature.capitalize()} by Supergenre')
    ax.set_xlabel('Mean Value')
    ax.legend(fontsize=8)

plt.suptitle('Mean Audio Feature Values by Supergenre Cluster', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```

This demonstrates at a minimum that the supergenre classifications are not *entirely* arbitrary, and do carry some amount of similarity since there are noticeable differences in some of these features.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler

audio_features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                  'instrumentalness', 'liveness', 'valence', 'tempo']
short_names = {
    'Rock, Metal & Punk Universe':               'Rock/Metal',
    'Electronic & Dance Music':                  'Electronic',
    'Pop, Entertainment & Global Pop':           'Pop',
    'Latin, Reggae & World Cultural Traditions': 'Latin/Reggae',
    'Roots, Classical & Introspective Music':    'Roots/Classical',
}

cluster_means = data.groupby('cluster')[audio_features].mean()
cluster_means.index = cluster_means.index.map(short_names)

scaler = StandardScaler()
cluster_means_scaled = pd.DataFrame(
    scaler.fit_transform(cluster_means),
    index=cluster_means.index,
    columns=cluster_means.columns
)

fig, ax = plt.subplots(figsize=(12, 4))
sns.heatmap(cluster_means_scaled, cmap='RdYlGn', center=0,
            annot=True, fmt='.2f', linewidths=0.5, ax=ax,
            cbar_kws={'label': 'Z-score (vs. dataset mean)'})
ax.set_title('Supergenre Audio Fingerprints\n(Z-scored mean per feature)', fontsize=13)
plt.tight_layout()
plt.show()
```

Similar to the conclusion above, there are noticeable differences per Supergenre for the audio features. Reducing the classes has greatly improved the ability to recognize distinctions between the classes!
