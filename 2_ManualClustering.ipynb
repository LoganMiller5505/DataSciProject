```python
import pandas as pd

data = pd.read_csv('data//1_preprocessed//preprocessed_dataset.csv')

# apply cluster mapping to the 'track_genre' column to create a new 'cluster' column
def map_genre_to_cluster5(genre):
    for cluster_name, genres in clusters_5.items():
        if genre in genres:
            return cluster_name
    return "Unknown Cluster"

data['cluster'] = data['track_genre'].apply(map_genre_to_cluster5)

unknown5 = data[data['cluster'] == "Unknown Cluster"]['track_genre'].unique()
```
