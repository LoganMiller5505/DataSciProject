```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('data//2_clustered//clustered_dataset.csv')

data['energy_acoustic_ratio']     = data['energy'] / (data['acousticness'] + 1e-6)
data['acoustic_calm']             = data['acousticness'] * (1 - data['energy'])
data['loudness_norm']             = (data['loudness'] - data['loudness'].min()) / (data['loudness'].max() - data['loudness'].min())
data['loudness_x_acousticness']   = data['loudness_norm'] * data['acousticness']
data['loudness_x_energy']         = data['loudness_norm'] * data['energy']
data['dance_energy_ratio']        = data['danceability'] / (data['energy'] + 1e-6)
data['tempo_x_energy']            = (data['tempo'] / 250) * data['energy']
data['energy_x_danceability']     = data['energy'] * data['danceability']

data.to_csv('data//3_engineered//engineered_dataset.csv', index=False)
```
