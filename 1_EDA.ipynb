# Initial Preprocessing Checks

## Check for Missing Data

```python
df.isnull().sum()
```

There is one song that has null values for artists, album name, and track name. Let's identify where this occurs.

```python
rows_with_missing_values = df[df.isnull().any(axis=1)]
rows_with_missing_values
```

Given that the popularity and duration are both 0 in addition to the missing artist, album name, and track name information, this is more than likely an error with the dataset that can be safely removed.

The missing entry has been deleted

## Check Duplicate Data

```python
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

There are 450 duplicate rows. Let's see how many duplicates there are for the tracks, to see if there are some tracks being duplicated many times or just a few duplicates for each track

```python
duplicate_row_counts = df[df.duplicated(keep=False)].groupby(list(df.columns)).size().reset_index(name='count')
duplicate_row_counts['count'].unique()
```

No track has more than 3 duplicates. There likely isn't some underlying trend to be maintained and it is a simple error, so we can remove these duplicates.

## Remove Duplicate Data

```python
df = df.drop_duplicates()
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

Calculate the mean, std, min, max, and quantiles for everything in numeric_columns

Some gathered insights from this:
- The mean and median track length is 3.8 minutes and 3.5 minutes, respectively, but the longest track is 87 minutes long!
- Many of the float audio features are right skewed with many entries close to or at 0 but just a few close to or at 1 (see "instrumentalness" and "speechiness" for good examples)

### Tempo Analysis

Let's act on one of these insights, analyzing where the tempo was 0 BPM and if that makes sense

```python
zero_tempo_rows = df[df['tempo'] == 0]
zero_tempo_rows
```

Let's see if there are some genres which more frequently have a tempo of 0 than others.

```python
genre_counts = zero_tempo_rows['track_genre'].value_counts()
print(genre_counts)
```

We can see that some of these tracks with a tempo of 0 are white noise (whose genre is typically categorized as "sleep"), so it makes sense that they wouldn't have a tempo. Others, however, including sountracks and guitar music, should have a listed tempo.

Unfortunately, there is no easy way to distinguish between white noise and standard tracks in this list since some of the genre distinctions are not clear (several of the tracks under "world-music" are actually "ambient" or "sleep" tracks, for example).

Due to this lack of clear distinction, they will be left in the dataset for now. However, should the overall hypothesis and objective in the future focus on specifically songs rather than including ambient tracks, then it would be reasonable to delete all tracks with a tempo of 0.

This is something to note for later as potentially actionable, but for now, we will continue analyzing the overall data.

# Genre Relationships

## Top and Bottom 10 Genres in Select Audio Features

```python
import matplotlib.pyplot as plt
import pandas as pd

features_to_plot = ['danceability', 'energy', 'acousticness', 'speechiness']
genre_means = df.groupby('track_genre')[features_to_plot].mean()

fig, axes = plt.subplots(1, 4, figsize=(25, 7))
for ax, feature in zip(axes, features_to_plot):
    top10 = genre_means[feature].nlargest(10)
    bot10 = genre_means[feature].nsmallest(10)
    combined = pd.concat([bot10, top10]).sort_values()
    colors = ['tomato' if g in top10.index else 'steelblue' for g in combined.index]
    ax.barh(combined.index, combined.values, color=colors)
    ax.axvline(genre_means[feature].mean(), color='black', linestyle='--', linewidth=1, label='Dataset mean')
    ax.set_title(f'{feature.capitalize()}: Top (red) & Bottom (blue) 10 Genres')
    ax.set_xlabel('Mean Value')
    ax.legend(fontsize=8)
plt.suptitle('Genres at the Extremes of Key Audio Features', y=1.01, fontsize=13)
plt.tight_layout()
plt.show()
```

There is a large difference between the top and bottom, suggesting at least inuitively that genres can be deduced from these audio features.

## Heatmap of audio features for all genres

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler

audio_features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                  'instrumentalness', 'liveness', 'valence', 'tempo']

genre_means = df.groupby('track_genre')[audio_features].mean()
scaler = StandardScaler()
genre_means_scaled = pd.DataFrame(
    scaler.fit_transform(genre_means),
    index=genre_means.index,
    columns=genre_means.columns
)

fig, ax = plt.subplots(figsize=(12, 30))
sns.heatmap(genre_means_scaled, cmap='RdYlGn', center=0,
            linewidths=0.3, ax=ax,
            cbar_kws={'label': 'Z-score (vs. dataset mean)'})
ax.set_title('Genre Audio Fingerprints (Z-scored mean per feature)', fontsize=13)
plt.tight_layout()
plt.show()
```

While many genres are the same, there are several that have particularly noticeable deviations in some features. Examples are:
- Comedy (very high in speechiness and liveness)
- Sleep (low in danceability, loudness, and tempo)
This aligns with both common understandings of these genres and some of the observations noted before (sleep having lots of zero tempo tracks, for example).

However, it is difficult to tell some of the "crowd" apart from one another. This is what our model will hopefully accomplish, though!
