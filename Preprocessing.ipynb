# Extreme Values
Analyze potentially extreme values to see if any adjustments are necessary

```python
# Calculate extreme values for each column. Use the following format:
'''
#### Identify Extreme Data Values

# Importing necessary libraries
import pandas as pd

# Load the data
data = pd.read_csv('vote.csv')

# Calculate the 5th and 95th percentiles of age
age_5th_percentile = data['age'].quantile(0.05)
age_95th_percentile = data['age'].quantile(0.95)

# Get the range of the oldest 5% and youngest 5% of respondents
oldest_5_percent_range = data[data['age'] >= age_95th_percentile]['age'].agg(['min', 'max'])
youngest_5_percent_range = data[data['age'] <= age_5th_percentile]['age'].agg(['min', 'max'])

# Count the number of respondents above 100 and below 18
above_100_count = data[data['age'] > 100].shape[0]
below_18_count = data[data['age'] < 18].shape[0]

print('Highest age of the lowest 5%:', age_5th_percentile)
print('Lowest age of the highest 5%:', age_95th_percentile)
print('Oldest 5% age range:', oldest_5_percent_range)
print('Youngest 5% age range:', youngest_5_percent_range)
print('Number of respondents above 100:', above_100_count)
print('Number of respondents below 18:', below_18_count)
'''

numeric_columns = raw_data.select_dtypes(include='number').columns

for column in numeric_columns:
    # Calculate the 5th and 95th percentiles
    lower_bound = raw_data[column].quantile(0.05)
    upper_bound = raw_data[column].quantile(0.95)
    
    # Get the range of the lowest 5% and highest 5% of values
    lowest_5_percent_range = raw_data[raw_data[column] <= lower_bound][column].agg(['min', 'max'])
    highest_5_percent_range = raw_data[raw_data[column] >= upper_bound][column].agg(['min', 'max'])\
    
    # Calculate number of values above the upper bound and below the lower bound
    above_upper_bound_count = raw_data[raw_data[column] > upper_bound].shape[0]
    below_lower_bound_count = raw_data[raw_data[column] < lower_bound].shape[0]

    # Calculate the percentage of values above the upper bound and below the lower bound
    total_count = raw_data.shape[0]
    above_upper_bound_percentage = (above_upper_bound_count / total_count) * 100
    below_lower_bound_percentage = (below_lower_bound_count / total_count) * 100
    
    print(f'Column: {column}')
    print(f'Lowest 5% range: {lowest_5_percent_range}')
    print(f'Highest 5% range: {highest_5_percent_range}')
    print(f'Number of values above upper bound: {above_upper_bound_count}')
    print(f'Number of values below lower bound: {below_lower_bound_count}')
    print(f'Percentage of values above upper bound: {above_upper_bound_percentage:.2f}%')
    print(f'Percentage of values below lower bound: {below_lower_bound_percentage:.2f}%')

    print()
```
